```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("/content/Churn_Modelling.csv")
df.info()
```

```python
#check how many values are there
print(df.isnull().sum())
```

delete the column or the row with the missing values


```python
#delete the colums with missing values

updated_df = df.dropna(axis=1)
updated_df.info()
```

```python
#delete the row with missing values

updated_df = df.dropna(axis=0)
updated_df.info()
```

```python
#replace the missing value with the mean
df['Age'].mean()
```

```python
df['Age'].median()
```

```python
updated_df = df
updated_df['Age']= updated_df['Age'].fillna(df['Age'].mean)
updated_df.info()
```

```python
updated_df1 = df
updated_df1['Age'] = updated_df1['Age'].fillna(df['Age'].median)
updated_df.info()
```
