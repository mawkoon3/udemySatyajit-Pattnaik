```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
%matplotlib inline
```

```python
telcom_data = pd.read_csv('Churn_Modelling.csv')
```

```python
telcom_data.head(5)
```

```python
telcom_new = telcom_data[['Geography','Gender','Exited']]
telcom_new.head(5)
```

```python
telcom_new.Exited.value_counts()
```

```python
telcom_new.Exited.value_counts()/100
```

```python
#univariate analysis
for i,predictor in enumerate (telcom_new.drop(columns=['Exited'])):
  plt.figure()
  sns.countplot(data=telcom_new, x=predictor, hue='Exited')
```

```python
#bivariate analysis
sns.histplot(data=telcom_new, x = 'Gender', hue='Geography',stat = 'count',multiple = 'dodge', shrink= 0.9)
```

```python
telecom_newtarget = telcom_new.drop(columns=['Exited'])
telecom_newtarget.head(5)
```

```python
sns.histplot(x='Gender', hue='Geography', data=telecom_newtarget, stat="count", multiple="dodge",shrink=.8)
```

```python
telcom_data.head(5)
```

```python
telcom_data['Age'].value_counts().sort_index(ascending=True).plot()
```

```python
telecom_data_onlyexited = telcom_data[telcom_data["Exited"]==1]
telecom_data_onlyexited.head(5)
```

```python
telecom_data_onlyexited['Age'].value_counts().sort_index(ascending=True).plot()
```

```python
Tot = sns.kdeplot(telcom_data.Age[(telcom_data["Exited"]==0)],
                  color="Red", shade=True)
Tot = sns.kdeplot(telcom_data.Age[(telcom_data["Exited"]==1)],
                  color="Blue", shade=True)
Tot.legend(["No Churn", "Churn"], loc='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Age')
Tot.set_title('Age by churn')
```

```python
sns.kdeplot(
    data=telcom_data,
    x="Balance",
    hue="Exited",
    fill=True,
    common_norm=False,
    palette="crest",
    alpha=0.2,
    linewidth=0.5
)

plt.xlabel("Balance")
plt.ylabel("Density")
plt.title("Balance by Churn")
plt.legend(["No Churn", "Churn"])
```

```python
sns.kdeplot(data=telcom_data,x='Tenure',hue="Exited",multiple="stack",fill=True)
plt.legend(["No Churn", "Churn"], loc='upper right')
plt.title('Tenure by churn')
```
