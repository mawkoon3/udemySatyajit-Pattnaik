OverSampling



```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler()
x_ros, y_ros = ros.fit_resample(X, y)
model_dt2 = DecisionTreeClassifier()
model_dt2.fit(x_ros, y_ros)
y_pred_dt2 = model_dt2.predict(x_test_scaled)
print(accuracy_score(y_test, y_pred_dt2))
from sklearn.metrics import classification_report
dt_report2 = classification_report(y_test, y_pred_dt2)
print(dt_report2)
```

```python
from imblearn.combine import SMOTEENN
smote_enn = SMOTEENN()
x_smote, y_smote = smote_enn.fit_resample(X, y)
model_dt3 = DecisionTreeClassifier()
model_dt3.fit(x_smote, y_smote)
y_pred_dt3 = model_dt3.predict(x_test_scaled)
print(accuracy_score(y_test, y_pred_dt3))
dt_report3 = classification_report(y_test, y_pred_dt3)
print(dt_report3)
```

```python
from imblearn.over_sampling import ADASYN
ada = ADASYN(random_state=42)
x_ada, y_ada = ada.fit_resample(X, y)
model_dt4 = DecisionTreeClassifier()
model_dt4.fit(x_ada, y_ada)
y_pred_dt4 = model_dt4.predict(x_test_scaled)
print(accuracy_score(y_test, y_pred_dt4))
dt_report4 = classification_report(y_test, y_pred_dt4)
print(dt_report4)
```

```python
from imblearn.combine import SMOTETomek
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Apply SMOTE only on training data
smtt = SMOTETomek(random_state=42)
X_train_smtt, y_train_smtt = smtt.fit_resample(x_train_scaled, y_train)

# Train model
model_dt5 = DecisionTreeClassifier(random_state=42)
model_dt5.fit(X_train_smtt, y_train_smtt)

# Predict on untouched test data
y_pred_dt5 = model_dt5.predict(x_test_scaled)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred_dt5))
print(classification_report(y_test, y_pred_dt5))
```

```python
from imblearn.under_sampling import AllKNN
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Initialize AllKNN
allknn = AllKNN()

# Apply AllKNN to resample the training data
x_allknn, y_allknn = allknn.fit_resample(x_train_scaled, y_train)

# Train a Decision Tree Classifier on the resampled data
model_dt_allknn = DecisionTreeClassifier(random_state=42)
model_dt_allknn.fit(x_allknn, y_allknn)

# Predict on the untouched test data
y_pred_allknn = model_dt_allknn.predict(x_test_scaled)

# Evaluate the model
print("Accuracy with AllKNN:", accuracy_score(y_test, y_pred_allknn))
print("\nClassification Report with AllKNN:\n", classification_report(y_test, y_pred_allknn))
```

```python
from imblearn.under_sampling import TomekLinks
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Initialize TomekLinks
tomek_links = TomekLinks(sampling_strategy='majority') # 'majority' means only majority class samples are removed

# Apply TomekLinks to resample the training data
x_tomek, y_tomek = tomek_links.fit_resample(x_train_scaled, y_train)

# Train a Decision Tree Classifier on the resampled data
model_dt_tomek = DecisionTreeClassifier(random_state=42)
model_dt_tomek.fit(x_tomek, y_tomek)

# Predict on the untouched test data
y_pred_tomek = model_dt_tomek.predict(x_test_scaled)

# Evaluate the model
print("Accuracy with TomekLinks:", accuracy_score(y_test, y_pred_tomek))
print("\nClassification Report with TomekLinks:\n", classification_report(y_test, y_pred_tomek))
```
