```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/mall.csv")
```

```python
df.head(5)
```

```python
x = df.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import MeanShift
ms = MeanShift(bandwidth=10)
ms.fit(x)
labels = ms.labels_
cluster_cenqters = ms.cluster_centers_
```

```python
n_clusters = len(np.unique(labels))
colors = ['red','blue','green','yellow','cyan','purple','pink']

for i in range(n_clusters):
  cluster = x[labels == i]
  plt.scatter(cluster[:,0], cluster[:,1], s=100, c=colors[i % len(colors)], label=f'Cluster {i+1}')
plt.title('Cluster of Clients')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score (0-100)')
plt.legend()
plt.show()
```
