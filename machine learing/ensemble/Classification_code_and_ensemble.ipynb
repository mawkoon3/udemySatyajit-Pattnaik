```python
from sklearn.ensemble import AdaBoostClassifier
model_ada = AdaBoostClassifier()
model_ada.fit(x_train_scaled, y_train)
y_pred_ada = model_ada.predict(x_test_scaled)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model_gb = GradientBoostingClassifier()
model_gb.fit(x_train_scaled, y_train)
y_pred_gb = model_gb.predict(x_test_scaled)
```

XG Boost

```python
from xgboost import XGBClassifier

# Convert 'No' to 0 and 'Yes' to 1 for y_train and y_test
y_train_encoded = np.where(y_train == 'Yes', 1, 0)
y_test_encoded = np.where(y_test == 'Yes', 1, 0)

model_xgb = XGBClassifier()
model_xgb.fit(x_train_scaled, y_train_encoded)
y_pred_xgb = model_xgb.predict(x_test_scaled)
```

```python
print("Accuracy of XGBoost:", accuracy_score(y_test_encoded, y_pred_xgb))
```
