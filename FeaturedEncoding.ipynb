###label encoding


```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['Gender'] = label_encoder.fit_transform(df['Gender'])
df.head()
```

```python
# Create new columns for one-hot encoding based on the existing 'Gender' column
# Assuming 0 represents Female and 1 represents Male from the previous label encoding
df['Gender_Female'] = (df['Gender'] == 0).astype(int)
df['Gender_Male'] = (df['Gender'] == 1).astype(int)
df.head()
```
